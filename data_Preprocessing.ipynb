## Converting categorical data

### 1. Gender

```python
Gender = pd.get_dummies(data['Gender']).astype(int)
```

```python
Gender
```

```python
Gender.dtypes
```

```python
data = pd.concat([data,Gender],axis=1)
```

```python
data.head()
```

### 2. EducationBackground

```python
data['EducationBackground'].value_counts()
```

```python
data['EducationBackground'] = data['EducationBackground'].map({'Life Sciences':5,'Medical':4,'Marketing':3,'Technical Degree':2,
                                                             'Other':1,'Human Resources':0})
```

```python
data.info()
```

### 3. MaritalStatus

```python
data['MaritalStatus'].value_counts()
```

```python
data['MaritalStatus'] = data['MaritalStatus'].map({'Married':3,'Single':2,'Divorced':1})
```

### 4. EmpDepartment

```python
data['EmpDepartment'].value_counts()
```

```python
data['EmpDepartment'] = data['EmpDepartment'].map({'Sales':6,'Development':5,'Research & Development':4,'Human Resources':3,
                                                   'Finance':2,'Data Science':1})
```

### 5. EmpJobRole

```python
data['EmpJobRole'].value_counts()
```

```python
data['EmpJobRole'] = data['EmpJobRole'].map({'Sales Executive':18,'Developer':17,'Manager R&D':16,'Research Scientist':15,
                                             'Sales Representative':14,'Laboratory Technician':13,'Senior Developer':12,
                                             'Manager':11,'Finance Manager':10,'Human Resources':9,'Technical Lead':8,
                                             'Manufacturing Director':7,'Healthcare Representative':6,'Data Scientist':5,
                                            'Research Director':4,'Business Analyst':3,'Senior Manager R&D':2,
                                             'Delivery Manager':1,'Technical Architect':0})
```

### 6. BusinessTravelFrequency

```python
data['BusinessTravelFrequency'].value_counts()
```

```python
data['BusinessTravelFrequency'] = data['BusinessTravelFrequency'].map({'Travel_Rarely':2,'Travel_Frequently':1,'Non-Travel':0})
```

### 7. OverTime

```python
data['OverTime'].value_counts()
```

```python
data['OverTime'] = data['OverTime'].map({'Yes':1,'No':0})
```

### 8. Attrition

```python
data['Attrition'] = data['Attrition'].map({'Yes':0,'No':1})
```
