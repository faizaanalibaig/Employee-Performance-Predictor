```python
# Importing basic libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel(r"C:\Users\Faizaan Ali Baig\Downloads\INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls")
```

## Univariate Analysis

```python
plt.figure(figsize=(45,45))
plotnumber = 1

for column in data:
    plt.subplot(7,4,plotnumber)
    sns.countplot(x= column, data = data)
    plt.ylabel('count',fontsize=25)
    plt.xlabel(column,fontsize=25)
    plotnumber+=1
plt.tight_layout()
plt.show()
```

## Bivariate Analysis

```python
plt.figure(figsize=(20,18))
plotnumber = 1

for column in data:
    plt.subplot(6,5,plotnumber)
    sns.barplot(y=column,x="PerformanceRating",data=data)
    plt.xlabel("PerformanceRating")
    plotnumber+=1

plt.tight_layout()
plt.show()
```

## Task 1 : Departmnetwise performence rating

```python
plt.figure(figsize=(15,6))

grouped = data.groupby(['EmpDepartment'])['PerformanceRating'].mean().sort_values(ascending=False)

sns.barplot(y=grouped.index,x=grouped.values)

plt.xlabel('Department_rating')
plt.ylabel('department')
plt.show()
```

```python
plt.figure(figsize=(8,8))

pl_no=1

for col in data['EmpDepartment'].unique():
    plt.subplot(3,2,pl_no)
    df2 = data[data['EmpDepartment']==col]["PerformanceRating"].value_counts()
    sns.barplot(x=df2.index,y=df2.values)
    plt.xlabel(col)
    pl_no+=1
plt.tight_layout()
plt.show()
```
